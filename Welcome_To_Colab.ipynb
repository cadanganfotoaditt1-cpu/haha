```
print("\n2. Array berisi nilai mahasiswa dan tampilkan nilai tertinggi:  ")
nilai_mahasiswa = a.array([85, 90, 78, 92, 88, 95, 80, 91, 89, 94, 87, 93, 86, 90, 91])
nilai_tertinggi = a.max(nilai_mahasiswa)
nilai_rata = a.average(nilai_mahasiswa)
nilai_terkecil = a.min(nilai_mahasiswa)

print(nilai_mahasiswa)
print("\nNilai tertinggi:", nilai_tertinggi)
print("\nNilai Rata:", nilai_rata)
print("\nNilai Terendah:", nilai_terkecil)
```
